```python
mask.target = mask.target.astype("str")
```

```python
train_image_generator = ImageDataGenerator(rescale=1./255)
```

```python
train_data_gen = train_image_generator.flow_from_dataframe(mask, x_col="image", y_col = "target", directory = "images_mask",
                                                           shuffle=True,
                                                           target_size=(200, 200),
                                                           class_mode='binary')

```

```python
test_data_gen = train_image_generator.flow_from_dataframe(ss, x_col="image", directory = "images_mask",
                                                           shuffle=False,
                                                           target_size=(200, 200),
                                                          class_mode = None
                                                           )
```

```python
model = Sequential([
    Conv2D(16, 3, padding='same', activation='relu', input_shape=(200, 200 ,3)),
    MaxPooling2D(),
    Conv2D(32, 3, padding='same', activation='relu'),
    MaxPooling2D(),
    Conv2D(64, 3, padding='same', activation='relu'),
    MaxPooling2D(),
    Flatten(),
    Dense(512, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(train_data_gen,
          steps_per_epoch = train_data_gen.n//train_data_gen.batch_size,
          epochs=20, verbose=1)
```
